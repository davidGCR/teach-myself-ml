```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

# Simple plot with default settings
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
# Instantiating defaults
# ------------------------------------

plt.figure(figsize=(6, 4), dpi=80)
plt.subplot(1, 1, 1)

plt.plot(X, C, color='blue', linewidth=2.0, linestyle='-')
plt.plot(X, S, color='green', linewidth=1.0, linestyle='-')

plt.xlim(-4.0, 4.0)
plt.xticks(np.linspace(-4, 4, 5, endpoint=True))
plt.ylim(-1.5, 1.5)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.show()
```

```python
# Display TICKS
# ------------------------------------
def draw_sin_cos_then_hold(caller=plt):
    """
    Draw sin cos plots, then hold it. Dont call plt.show() immediately. 
    Used for different experiments without many times of copy/past.
    """
    caller.plot(X, C, color='blue', linewidth=1.0, linestyle='-', label='cosine')
    caller.plot(X, S, color='red', linewidth=1.0, linestyle='-', label='sine')

draw_sin_cos_then_hold()

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
# Display TICKS with ARBITRARY TEXT
# ------------------------------------
# The xticks in the previous chart look not natural.
# Should display it as `pi/2` instead of `1.571`
# plt.xticks(locations, LABELS)
draw_sin_cos_then_hold()

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],   # Locations 
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])  # Labels

# [Question] What is `r` here?
plt.yticks([-1, 0, 1])
plt.show()
```

```python
# Moving spines
# ------------------------------------
# Spines are lines connecting axis tick marks and noting the boundaries of data area.
draw_sin_cos_then_hold()

def shift_spines_to_center(caller=plt):
    ax = caller.gca()  # gca = `get current axis`
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data', 0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data', 0))
    
shift_spines_to_center()
```

```python
# Adding legends
# ------------------------------------
draw_sin_cos_then_hold()

plt.legend(['cosine', 'sine'], loc='upper left')
plt.show()

# or... declare label it in plot()
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-", label="cosine")
plt.plot(X, S, color="red",  linewidth=1.0, linestyle="-", label="sine")
plt.legend(loc='upper left')
plt.show()
```

```python
# Annotate some points
# ------------------------------------
draw_sin_cos_then_hold()
shift_spines_to_center()

t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=1.5, linestyle='--')
plt.scatter([t, ], [np.cos(t), ], 50, color='blue') # Blue dot
plt.plot([t, t], [0, np.sin(t)], color='red', linewidth=1.5, linestyle='--')
plt.scatter([t, ], [np.sin(t), ], 50, color='red')  # Red dot

# [Question] These params are extremely hard to remember!!!
# and argument labels are not explicit!!!
# Need a WRAPPER!
plt.annotate(
    r'$sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
    xy=(t, np.sin(t)), xycoords='data',
    xytext=(10, 30), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2")) # Really ADVANCED!
plt.annotate(
    r'$cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
    xy=(t, np.cos(t)), xycoords='data',
    xytext=(-90, -50), textcoords='offset points', fontsize=16,
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.legend(loc='upper left')
plt.show()
```
